```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r"D:\HENRY\M1\bquilla1\DATASET_QUILLA_DIA1.csv", header=0)
df1
```

```python
print(df1.dtypes)
```

```python
# Nombre de la columna que deseas convertir
columna_a_convertir = df1.columns[0]

# Convertir la columna de "object" a "float"
df1[columna_a_convertir] = pd.to_numeric(df1[columna_a_convertir], errors="coerce")
df1[columna_a_convertir] = df1[columna_a_convertir].astype(float)
```

```python
columna_a_convertir = df1.columns[3]

# Convertir la columna de "object" a "float"
df1[columna_a_convertir] = pd.to_numeric(df1[columna_a_convertir], errors="coerce")
df1[columna_a_convertir] = df1[columna_a_convertir].astype(float)
```

```python
df1.sort_values(by='MINUTO ', ascending=True , inplace = True )
```

```python
sns.set(style="whitegrid")

# Crear un gráfico de barras utilizando la columna "MINUTO" como base
plt.figure(figsize=(10, 6))  # Tamaño de la figura

sns.histplot(data=df1, x="MINUTO ", bins=20, kde=True)  # Histograma con densidad

# Agregar título y etiquetas a los ejes
plt.title("Distribución de registros por minuto")
plt.xlabel("Minuto")
plt.ylabel("Frecuencia")

# Mostrar el gráfico
plt.show()
```
